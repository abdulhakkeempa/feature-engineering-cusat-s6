```python
# Import necessary libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris

# Load Iris dataset
iris = load_iris()
iris_df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
iris_df['target'] = iris.target

# Display summary statistics
summary_stats = iris_df.describe()
print(summary_stats)

# Bivariate Analysis
sns.boxplot(x='target', y='sepal length (cm)', data=iris_df)
plt.show()

# Multivariate Analysis
sns.heatmap(iris_df.corr(), annot=True, cmap='coolwarm', linewidths=0.5)
plt.show()
```

```python
# Display Unique Values and Missing Values
unique_values = iris_df.nunique()
missing_values = iris_df.isnull().sum()

print("Unique Values in Each Column:")
print(unique_values)

print("\nNumber of Missing Values in Each Column:")
print(missing_values)
```

```python
# Univariate Analysis
plt.figure(figsize=(12, 6))

# Visualizing Sepal Length
plt.subplot(1, 2, 1)
sns.histplot(iris_df['sepal length (cm)'], kde=True, color='skyblue')
plt.title('Distribution of Sepal Length')

# Visualizing Petal Length
plt.subplot(1, 2, 2)
sns.histplot(iris_df['petal length (cm)'], kde=True, color='salmon')
plt.title('Distribution of Petal Length')

plt.tight_layout()
plt.show()
```
